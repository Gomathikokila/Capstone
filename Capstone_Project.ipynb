# Analysis of Literacy Rate vs Total Crimes.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv(r'https://raw.githubusercontent.com/Gomathikokila/Capstone/main/2014%20Total%20crimes.csv')
df.head()
```

```python
df.columns
```

```python
df.columns.get_loc("Total Cognizable IPC crimes")
```

```python
df.columns.get_loc("Total_Theft_Value of property stolen")
```

```python
plt.figure(figsize=(20,15),facecolor='red')
plotnumber=1
for column in range(165,177):
    col_name= df.columns[column]
    if plotnumber <= 13:
        ax=plt.subplot(2,6,plotnumber)
        plt.scatter(df['Avrage Literacy rate'],df[col_name])
        plt.xlabel('Avrage Literacy rate',fontsize=20)
        plt.ylabel(col_name,fontsize=20)
    plotnumber+=1
plt.tight_layout()
```

We can clearly see that , there is no relation between literacy rate and all types of crime.

# Unsupervised ML (Clustering)

```python
from sklearn import datasets
from sklearn.cluster import KMeans

import warnings
warnings.filterwarnings('ignore')
```

# IPC Crimes

```python
df=pd.read_csv(r"https://raw.githubusercontent.com/Gomathikokila/Capstone/main/2014%20Total%20crimes.csv")
df.head()
```

```python
km=KMeans(n_clusters=3)
km
```

```python
y_pred=km.fit_predict(df[['Total Cognizable IPC crimes']])
y_pred
```

```python
df['cluster']=y_pred
df
```

```python
plt.figure(figsize=(20,20),facecolor='yellow')
plt.scatter(df['State'], df['Total Cognizable IPC crimes'], c=df['cluster'], cmap='viridis', marker='o', edgecolor='black')
plt.title('K-Means Clustering')
plt.xlabel('State')
plt.ylabel('IPC Crimes')
plt.show()
```

## Peaceful and Sensitive Places

```python
Peaceful_IPC=df[df.cluster==0]
Moderate_IPC=df[df.cluster==1]
Sensitive_IPC=df[df.cluster==2]
```

```python
Peaceful_IPC
```

```python
Moderate_IPC
```

```python
Sensitive_IPC
```

IN IPC crimes, Andra,Assam,Chhattisgarh,Gujarat,Haryana,Karnataka,Odisha,Telangana states are very sentive. More IPC crimes are getting register in those states.


Bihar, Kerala,MP,Maharashtra,Rajasthan,Tamil Nadu,UP,West Bengal are comming under peaceful place in the IPC crimes
